```python
import skimage
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Generate Random Image
random_image = np.random.random((500,500))
plt.imshow(random_image, cmap ='gray')  #Show stuff
plt.colorbar()
```

```python
from skimage import data
coins = data.coins()

print("Type:",type(coins))
print("dtype:",coins.dtype)
print("Type:",coins.shape)
```

```python
plt.imshow(coins, cmap ='gray')
```

```python
cat = data.chelsea()
print(cat.shape)
#3 chanels: Red Blue Green
print(cat.max(),cat.min())
plt.imshow(cat[:,:,0],cmap='gray') #only print the black & white channel
plt.colorbar()
```

```python
red_cat = cat[:,:,0]/225 #Normalize
```

```python
red_cat_shape = red_cat.shape
noise = np.random.random(red_cat_shape)*0.75
plt.imshow(red_cat+noise, cmap ='gray')
```

```python
plt.imshow(red_cat*0.1, cmap ='gray',vmin=0,vmax=1)
```

```python
plt.hist(red_cat)
```

```python
vals, bins = np.histogram(red_cat)
plt.plot(bins[:-1],vals)
#Bunch of bright pictures
```

```python
vals, bins = np.histogram(coins)
plt.plot(bins[:-1],vals)
```

```python
#Given the above hist, bright points are mostly >40
plt.imshow(coins>40,cmap='gray')
```

```python
#Blurr image
plt.imshow(skimage.filters.gaussian(coins, sigma=5), cmap='gray') #Create a blurry image
```

```python
bcoins = skimage.filters.gaussian(coins, sigma=5)
xbcoins = bcoins[1:,:] - bcoins [:-1,:] 
plt.imshow(xbcoins,cmap='gray')
plt.colorbar()
```

```python
xbcoins = (bcoins[1:,1:] - bcoins [:-1,1:])**2 + (bcoins[1:,1:] - bcoins [1:,:-1])**2
```

```python
plt.imshow(xbcoins>0.0005,cmap='gray')
```
