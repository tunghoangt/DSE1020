```python
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
os.getcwd()
```

# Load and clean

```python
x = np.load('Sign-language-digits-dataset/X.npy') #Images data
```

```python
#There are 2062 images of sign language digits from 0-9, each is a 64 x 64 dimension
x.shape
```

```python
plt.imshow(x[2]) #samples
```

```python
#Reshape the data of each image to 1 row so that PCA can understand it
data = x.reshape(len(x),-1)
data.shape
```

# Dimension reduction

```python
#Reduce the dimension from 2062 (64x64) to 10 components
n = 10
model = PCA(n_components = n)
model.fit(data)

#Extract the component data from the model:
components = model.components_
components.shape
```

# Visualization

```python
#The 10 components are the 10 features that are represent in the 2062 images.
feature_img = components[:n]

#Let's see what the features are:
plt.figure(figsize=(5,5))
for img in feature_img:
    plt.imshow(img.reshape(64,64))
```

### It seems like the model determine fingers are the most important features
