```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

import os
os.environ["PROJ_LIB"] = "/anaconda/share/proj"; #locate folder to epsg
from mpl_toolkits.basemap import Basemap
```

# Load and Clean

```python
df = pd.read_csv("Crimes_-_2001_to_present.csv")
df.dropna(inplace = True)
```

```python
area_of_interest = ['ASSAULT', 'BATTERY', 'THEFT']

df_map = df[['Date', 'Primary Type', 'Latitude', 'Longitude']]
df_map = df_map[df_map['Primary Type'].isin(area_of_interest)]
```

```python
#Read in datetime format
df_map['Date'] = pd.to_datetime(df_map['Date'], format="%m/%d/%Y %I:%M:%S %p")

# # filter year only on 2017
df_map = df_map[(df_map['Date'] > '2017-1-1') & (df_map['Date'] <= '2017-12-30')]

df_map.index = df_map['Date']

# filter on rush hours
df_map_travel = df_map.between_time('4:00', '7:00')
```

```python
# Chicago coordinates.
lowerlon = -88.2 
upperlon = -87.2
lowerlat = 41.62
upperlat = 42.05

plt.figure(figsize=(20,10))

m = Basemap(
    llcrnrlon=lowerlon,
    llcrnrlat=lowerlat,
    urcrnrlon=upperlon,
    urcrnrlat=upperlat,
    resolution='c',
    projection='lcc',
    lat_0=lowerlat,
    lat_1=upperlat,
    lon_0=lowerlon,
    lon_1=upperlon
    )

shape_path = 'Boundaries - Community Areas (current)/geo_export_69adb049-3d95-4cf1-968b-64b7d8e1e850'

shp_info = shp_area = m.readshapefile(shape_path, name='Boundaries', drawbounds=True)
plt.gca().axis("on")

  
lons, lats = m(df_map_travel['Longitude'].values, df_map_travel['Latitude'].values)
m.scatter(lons, lats, marker = 'o', color='r', s=1,)


plt.show()
```

# Clustering

```python
#To 3 clusters:
labels_6 = KMeans(3, random_state=0).fit_predict(df_map_travel[['Longitude', 'Latitude']])

#To 5 clusters
labels_10 = KMeans(5, random_state=0).fit_predict(df_map_travel[['Longitude', 'Latitude']])
```

# Visualization

```python
#Label 6

shp_info = shp_area = m.readshapefile(shape_path, name='Boundaries', drawbounds=True)
plt.gca().axis("on")

plt.figure(figsize=(20,10))
m.scatter(lons, lats, marker = 'o', c=labels_6, cmap='viridis')

plt.show()
```

```python
#Label 10

shp_info = shp_area = m.readshapefile(shape_path, name='Boundaries', drawbounds=True)
plt.gca().axis("on")

plt.figure(figsize=(20,10))
m.scatter(lons, lats, marker = 'o', c=labels_10, cmap='viridis')

plt.show()
```
