# Processing Images

Several libraries you could use

* numpy (really basic stuff)
* pillow ( formally PIL this has lots of image formats)
* opencv (Computer vision library wrapping C/C++)
* scikit-image (simplier and less features than opencv ... focused on image processing)

```python
import skimage
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
random_image = np.random.random((500,500))
plt.imshow(random_image, cmap='gray')
plt.colorbar()
```

```python
from skimage import data
coins = data.coins()

print("Type: ", type(coins))
print('dtype', coins.dtype)
print('shape', coins.shape)
```

```python
plt.imshow(coins)
```

```python
plt.imshow(coins,cmap='gray')
```

```python
cat = data.chelsea()
print(cat.shape)
print(cat.max(),cat.min())
plt.imshow(cat[:,:,0],cmap='gray')
plt.colorbar()
```

```python
red_cat = cat[:,:,0]/255.0
```

```python
red_cat_shape = red_cat.shape
noise = np.random.random(red_cat_shape)*(0.3)

plt.imshow(red_cat+noise,cmap='gray')
```

```python
plt.imshow(red_cat+noise*3,cmap='gray',vmin=0,vmax=1)
```

```python
plt.imshow(red_cat,cmap='gray',vmin=0,vmax=1)
```

```python
plt.imshow(red_cat*0.5,cmap='gray',vmin=0,vmax=1)
```

```python
plt.imshow(skimage.filters.gaussian(coins,sigma=5),cmap='gray')
```

```python
bcoins = skimage.filters.gaussian(coins,sigma=5)
xbcoins = bcoins[1:,:]-bcoins[:-1,:]
plt.imshow(xbcoins,cmap='gray')
```

```python
xbcoins = (bcoins[1:,1:]-bcoins[:-1,1:])**2 + (bcoins[1:,1:]-bcoins[1:,:-1])**2
plt.imshow(xbcoins,cmap='gray')
plt.colorbar()
```

```python
plt.imshow(xbcoins>0.0005,cmap='gray')
```
