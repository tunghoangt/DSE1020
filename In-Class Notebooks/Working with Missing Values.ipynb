# Three strategies for dealing with missing data

1. One thing we could do is drop any row that has any missing values. We did this above.
2. We could replace any missing value with zero.
3. We could replace any missing value with the mean or median.

Lets replace with zero and see what happens.

```python
company_features_df_zero = pivot_df[pop_ind].copy()
```

```python
company_features_df_zero.shape
```

```python
company_features_df_zero.isnull().any()
```

There are nulls everywhere. 

```python
company_features_df_zero.info()
```

Only about 6K to 4K not null. Lets fill with zeros.

```python
company_features_df_zero.fillna(0,inplace=True)
```

```python
company_features_df_zero.info()
```

All non-zero now. Unfortunatly we might care that we just changed the means:


```python
 pivot_df[pop_ind].mean(axis=0)
```

```python
company_features_df_zero.mean(axis=0)
```

Instead lets use the means to fill in the values and see what happens then.

```python
company_features_df_mean_fill = pivot_df[pop_ind].copy()
```

```python
fill_values = dict(zip(pop_ind,pivot_df[pop_ind].mean(axis=0).values))
```

```python
company_features_df_mean_fill.fillna(fill_values, inplace=True)
```

```python
company_features_df_mean_fill.mean(axis=0)
```

```python
company_features_df_mean_fill.info()
```

So here we replace the missing value with the mean of that attribute. In this way the mean of the attirbute over the data does not change.
