## Data Set Mini Task (1)

```python
#Unsupervised learning
#Load the digit dataset in SKlearn. Standardize the data and split into test/train split using test_size = 0.25 
# and random_state = 42. Partition the data using KMeans into 10 clusters. Print the Crosstab table.

#Import neccessary package
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cross_validation import train_test_split
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np

#Load the digit data:
digits = datasets.load_digits()

#Standardize the digits data with StandardScaler()
scaler = StandardScaler()
data = scaler.fit_transform(digits.data)

# Split the `digits` data into training and test sets:
X_train, X_test, y_train, y_test = train_test_split(data, digits.target, test_size=0.25, random_state=42)

#Create an instance of KMeans with 10 cluster and a random state of 42. Assign to 'model'
model = KMeans(n_clusters = 10, random_state=42)

#Fit 'model' with Training set and then predict using Test data set, assigning results to 'y_pred'
model.fit(X_train)
y_pred = model.predict(X_test)

#Aligning predicted and actual numbers by creating a Data frame from a Dictionary (keys: 'predicted numbers', 'actual numbers')
df = pd.DataFrame({'predicted numbers': y_pred, 'actual numbers': y_test})

#Print out Crosstab table:
ct = pd.crosstab(df['predicted numbers'], df['actual numbers'])
print(ct)
```

### Question 2: PCA Analysis

```python

#Import neccessary package
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
np.random.seed(42)

#Setup dataset X that has 2 columns: Experience & Salary
n_samples = 100
experience = np.random.normal(size=n_samples)
salary = 1500 + experience + np.random.normal(size=n_samples, scale=.5)
X = pd.DataFrame({'experience':experience, 'salary':salary})


#Set up instance of PCA with 2 components:
pca = PCA(n_components=2)

#Fit PCA with X
pca.fit(X)

#Print "explained variance ratio" attributes
print(pca.explained_variance_ratio_)

#Transform X
PC = pca.transform(X)

#Plot a scatterplot of Experience and Salary
plt.subplot(121)
plt.scatter(X['experience'],X['salary'])

#Label the plot
plt.xlabel("Experience"); plt.ylabel("Salary")

#Plot 2 components of PCA to shows decorrelation of data
plt.subplot(122)
plt.scatter(PC[:, 0], PC[:, 1])

#Labeling:
plt.xlabel("PC1 (var=%.2f)" % pca.explained_variance_ratio_[0])
plt.ylabel("PC2 (var=%.2f)" % pca.explained_variance_ratio_[1])
plt.axis('equal')
plt.tight_layout()
```

### Question 1: Basic data manipulation with Pandas


```python
#Import neccessary package
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
columns = ['name', 'age', 'gender', 'job']
df1 = pd.DataFrame([['alice', 19, "F", "student"], ['john', 26, "M", "student"]], columns=columns)
df2 = pd.DataFrame([['eric', 22, "M", "student"], ['paul', 58, "F", "manager"]], columns=columns)
df3 = pd.DataFrame(dict(name=['peter', 'julie'], age=[33, 44], gender=['M', 'F'], job=['engineer', 'scientist']))
df4 = pd.DataFrame(dict(name=['alice','john','eric','julie'],height=[165,180,175,171]))

#Print to verify that the 3 dataframe have the same structure
print(df1)
print()
print(df2)
print()
print(df3)
print()

#Concatenate the first 3 dataframes:
df = pd.concat([df1,df2,df3])
print(df)
print()

#Join df with df4 on intersection of keys ("name"):
df_merged = pd.merge(df4,df, on = "name")
print(df_merged)
print()

#Union df and df4:
df_union = pd.merge(df,df4, on ="name", how='outer')
print(df_union)

#select 'job' and 'gender' of people in df_union that are younger than 21 or older than 50
df_union[(df_union.age < 21) | (df_union.age > 50)][['job','gender']]

#Sort df_union by age, in descending order:
df_union.sort_values(by='age', ascending=False)
```

### Question 3: Finding the most common word

```python
#Import neccessary package
%matplotlib inline
from bs4 import BeautifulSoup as BS
import requests
import nltk
import re

#Request text for processing:
request = requests.get('http://www.gutenberg.org/files/1661/1661-h/1661-h.htm')

#Process the HTML to get the text from the body:
html = request.text

#Create a BeautifulSoup object from the HTML:
soup = BS(html,"html5lib")

#Look at the title of the book:
soup.title

#get the text content of the book:
text = soup.get_text()

#Find all words in the book and print the first 10:
tokens = re.findall('\w+', text)
tokens[:10]

#Make all items in the token list lower case and append to an empty list "words"
words = []
for word in tokens:
    words.append(word.lower())
    
#Get English stopwords:
sw = nltk.corpus.stopwords.words('english')

#Create new list words_ns and add to it all words in tokens that are not in SW:
words_ns = []
for word in words:
    if word not in sw:
        words_ns.append(word)

#Create frequency distribution
fd = nltk.FreqDist(words_ns)

# #Plot most 10 common words:
fd.plot(10)
```

### Question 4: Basic exploratory data analysis:

```python
#Importing the neccessary package:
import pandas as pd
import seaborn as sns
import numpy as np

#Read in the Nobel Prize Data:
nobel = pd.read_csv('nobel.csv')

#Take a look at the 'full_name' and 'prize' of the first 10 winners:
print(nobel[['full_name','prize']].head(10))

# Display the number of Nobel Prizes handed out during the period
print(len(nobel))

# Display the number of prizes won by male and female recipients:
print(nobel["sex"].value_counts())

# Display the number of prizes won by the top 10 nationalities.
print(nobel["birth_country"].value_counts().head(10))

# Picking out the 'year', full_name', and 'prize' of the first woman to win a Nobel Prize
df1 = nobel[nobel["sex"] == "Female"]
print(df1.nsmallest(1, "year")[['year','full_name','prize']])

#Converting birth_date from String to datetime
nobel['birth_date'] = pd.to_datetime(nobel["birth_date"])

# Calculating the age of Nobel Prize winners and assign values to column 'age'
nobel['age'] = nobel["year"] - nobel["birth_date"].dt.year

# The oldest winner of a Nobel Prize as of 2016
display(nobel.nlargest(1, "age"))

# The youngest winner of a Nobel Prize as of 2016
nobel.nsmallest(1, "age")
```

### Question 5: Image Processing

```python
#Importing the neccessary package
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from PIL import Image

# open the image 'CCNY.jpg'
img = Image.open('CCNY.jpg')

# Get the image size
img_size = img.size
print("the image size is: {}".format(img_size))

#Display the image
display(img)

# rotate the image by 45 degrees & display
img_rotated = img.rotate(45, expand = 25)
display(img_rotated)

# Turn our image object into a NumPy array
img_data = np.array(img)

# get the shape of the resulting array
img_data_shape = img_data.shape

print("The NumPy array has the shape: {}".format(img_data_shape))

# plot the red channel
plt.imshow(img_data[:, :, 0], cmap=plt.cm.Reds_r)
plt.show()

# plot the green channel
plt.imshow(img_data[:, :, 1], cmap=plt.cm.Greens_r)
plt.show()

# plot the blue channel
plt.imshow(img_data[:, :, 2], cmap=plt.cm.Blues_r)
plt.show()

# convert honey to grayscale
img_bw = img.convert("L")

# convert the image to a NumPy array
img_bw_arr = np.array(img_bw)

# plot the array using matplotlib
plt.imshow(img_bw_arr, cmap=plt.cm.gray)
plt.show()
```
