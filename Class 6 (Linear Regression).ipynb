```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
num_points = 100
np.random.seed(42)
xs = np.random.random((num_points,))
noise_level = 0.1
es = np.random.rand(num_points)*noise_level
m = 2.718
b = 3.141
ys = xs*m + b + es
```

```python
fig,ax = plt.subplots()
ax.scatter(xs,ys)
ax.set_xlabel('Variable $x$')
ax.set_ylabel('Variable $y$')
ax.set_title('Synthetic linear data')
```

```python
fig,ax = plt.subplots()
ax.scatter(xs,ys,color='b')
ax.plot([xs.min(),xs.max()],[xs.min()*m + b,xs.max()*m + b],'r') #add synthetic line
ax.set_xlabel('Variable $x$')
ax.set_ylabel('Variable $y$')
ax.set_title('Synthetic linear data')
```

## Finding the Linear Regression

```python
X = np.hstack([xs.reshape((len(xs),1)),np.ones((len(xs),1))])
X.shape
```

```python
Y = ys.reshape((len(ys),1))
Y.shape
```

```python
Chat = np.dot(np.dot(np.linalg.inv(np.dot(X.T, X)),X.T),Y) #times, and then inverse, then multiplied by inverse
Chat
yshat = np.dot(X,Chat)
```

```python
fig,ax = plt.subplots()
ax.scatter(xs,ys,color='b',s=10)
ax.scatter(xs,yshat,color='y',s=10)
ax.plot([xs.min(),xs.max()],[xs.min()*m + b,xs.max()*m + b],'r') #add synthetic line
ax.plot(np.vstack([xs,xs]),np.vstack([np.squeeze(ys),np.squeeze(yshat)]),color='m')
ax.plot([xs.min(),xs.max()],[xs.min()*Chat[0][0] + Chat[1][0],xs.max()*Chat[0][0] + Chat[1][0]],'c')
ax.set_xlabel('Variable $x$')
ax.set_ylabel('Variable $y$')
ax.set_title('Synthetic linear data')
```

### Multivariable Linear Regression

```python
from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
print(boston.feature_names)
```

```python
#Set up dataset as DF
bos = pd.DataFrame(boston.data)
bos.columns = boston.feature_names
bos['Price'] = boston.target
bos.head()
```

```python
lm = LinearRegression()
lm.fit(bos.iloc[:,0:-1],bos.iloc[:,-1])
y_pred = lm.predict(bos.iloc[:,0:-1])
```
