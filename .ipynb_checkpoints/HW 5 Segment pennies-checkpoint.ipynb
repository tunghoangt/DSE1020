```python
import cv2
import numpy as np

coinz = cv2.imread("pennies.png", 0)
coinz = cv2.medianBlur(coinz,19)
cimg = cv2.cvtColor(coinz,cv2.COLOR_GRAY2BGR)

circles = cv2.HoughCircles(coinz,cv2.HOUGH_GRADIENT,0.5,50,
                            param1=250,param2=30,minRadius=0,maxRadius=0)

circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    # draw the outer circle
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)

plt.imshow(cimg)
```
