```python
from bs4 import BeautifulSoup as BS
```

```python
soup = BS(req.text,'html.parser')
```

```python
all_p = soup.find_all('p')
```

```python
len(all_p)
```

```python
all_tables = soup.find_all('table')
```

```python
len(all_tables)
```

```python
all_tables[0]
```

```python
tables_class = [table['class'] for table in all_tables]
tables_class
```

```python
wikitables = [table for table in all_tables if 'wikitable' in table['class']]
```

```python
len(wikitables)
```

```python
from IPython.display import HTML
```

```python
HTML(str(wikitables[0]))
```

```python
HTML(str(wikitables[1]))
```

```python
rows = [row for row in wikitables[1].find_all('tr')]
```

```python
rows[0].find_all('th')
```

```python
colheads = [col.get_text() for col in rows[0].find_all('th')]
colheads
```

```python
colheads = [col.get_text().strip().lower() for col in rows[0].find_all('th')]
colheads
```

```python
colheads = [col.get_text().strip().lower().replace(' ', '_').replace('.','') for col in rows[0].find_all('th')]
colheads
```

```python
data_rows = rows[1:]
data_rows[0]
```

```python
value_rows = [ [ col.get_text() for col in row.find_all('td') ] for row in rows[1:]]
value_rows
```

```python
value_series =list(zip(*value_rows))
value_series
```

```python
data_dict = dict(zip(colheads,value_series))
data_dict
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data_dict)
```

```python
df.head()
```

```python
df.tail()
```

```python
bycollege = df.groupby('college')
```

```python
size_table = bycollege.size()
size_table
```
