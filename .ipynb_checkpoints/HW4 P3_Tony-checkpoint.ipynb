# PART 2: Data story

```python
gridsize = (3,2)
fig = plt.figure(figsize = (12,8))
ax1 = plt.subplot2grid(gridsize, (0,0))
ax2 = plt.subplot2grid(gridsize, (0,1))
ax3 = plt.subplot2grid(gridsize, (1,0))
ax4 = plt.subplot2grid(gridsize, (1,1))
ax5 = plt.subplot2grid(gridsize, (2,0))

ax1.plot(labor['Trade'])
ax1.set_title('Trade employment')

ax2.plot(labor['Manufacturing'])
ax2.set_title('Manufacturing employment')

ax3.plot(labor['Education'])
ax3.set_title('Education employment')

ax4.plot(labor['Hospitality'])
ax4.set_title('Hospitality employment')

ax5.plot(labor['Finance'])
ax5.set_title('Finance employment')

plt.tight_layout()
```

#### Insight 1: 
We would expect that the level of employments for all the economic sectors would drop during the financial crisis. However, by looking at the graphs, we notice that Education is the exception as its level of employment has been increasing steadily since 2008. This could be due to various support programs introduced by the Obama administration when Obama first took the office in 2008.

#### Insight 2: 
Manufacturing employment seems to take the biggest hit as its current level of employment hasn't resume back to the pre-crisis level. Meanwhile, sectors like Hospitality, Trade, and even Finance seems to have recovered well with the level of employment exceeds pre-crisis level

```python
growth_MF = (labor['Manufacturing'][120]/labor['Manufacturing'][20] - 1)* 100

growth_Tr = (labor['Trade'][120]/labor['Trade'][20] - 1)* 100

growth_F = (labor['Finance'][120]/labor['Finance'][20] - 1)* 100

growth_Ed = (labor['Education'][120]/labor['Education'][20] - 1)* 100

growth_H = (labor['Hospitality'][120]/labor['Hospitality'][20] - 1)* 100

labels = ('Manufacturing','Trade','Finance','Education','Hospitality')
growth = np.array([growth_MF,growth_Tr,growth_F,growth_Ed,growth_H])
```

```python
fig,ax = plt.subplots()
ax.bar(labels,growth,align='center')
for i in range(5):
    plt.text(x=i-0.3,y = growth[i]+0.3, s = str(round(growth[i],2)) + '%', size = 10)
plt.subplots_adjust(bottom= 1, top = 2.5)
plt.title('Growth rate of employment since January 2010')
plt.ylabel('Percentage of growth')
plt.show()
```

#### Insight 3:
President Obama declared on January 27, 2010 that "the markets are now stabilized, and we've recovered most of the money we spent on the banks", marking the end of the financial crisis. Looking at employment data from January 2010 to now, we can see that the Hospitality sector is performing best with the highest jobs growth over 8 years, followed by Education sector. Manufacturing, as expected, is lagging behind all of them with a modest growth rate of 8.52%.
